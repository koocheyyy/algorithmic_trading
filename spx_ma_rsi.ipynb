```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Load SP500 tickers from CSV
tickers_df = pd.read_csv('spx.csv')
tickers = tickers_df['Symbol'].tolist()

# Define the date range for the past 3 years
start_date = '2021-08-13'
end_date = '2024-08-13'

# Fetch historical data for each ticker
def fetch_data(tickers, start, end):
    data = {}
    for ticker in tickers:
        try:
            data[ticker] = yf.download(ticker, start=start, end=end)
        except Exception as e:
            print(f"Error fetching data for {ticker}: {e}")
    return data

data = fetch_data(tickers, start_date, end_date)
```

```python
# Calculate RSI
def calculate_rsi(df, period=14):
    delta = df['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Calculate moving averages
def calculate_moving_averages(df, short_window=50, long_window=200):
    short_mavg = df['Close'].rolling(window=short_window, min_periods=1).mean()
    long_mavg = df['Close'].rolling(window=long_window, min_periods=1).mean()
    return short_mavg, long_mavg

# Mean reversion strategy
def mean_reversion_strategy(df):
    df['RSI'] = calculate_rsi(df)
    df['Short_MA'], df['Long_MA'] = calculate_moving_averages(df)
    
    # Buy signal: RSI < 30 and price < Short_MA
    df['Signal'] = 0
    df.loc[(df['RSI'] < 30) & (df['Close'] < df['Short_MA']), 'Signal'] = 1
    
    # Sell signal: RSI > 70 and price > Short_MA
    df.loc[(df['RSI'] > 70) & (df['Close'] > df['Short_MA']), 'Signal'] = -1
    
    df['Position'] = df['Signal'].diff().fillna(0)
    
    # Calculate returns
    df['Daily_Return'] = df['Close'].pct_change()
    df['Strategy_Return'] = df['Daily_Return'] * df['Signal'].shift(1)
    df['Cumulative_Return'] = (1 + df['Strategy_Return']).cumprod()
    
    return df

# Apply strategy to each stock
strategy_results = {}
for ticker, df in data.items():
    strategy_results[ticker] = mean_reversion_strategy(df)
```

```python
# Initialize portfolio value
initial_portfolio_value = 1000

# Create a DataFrame to hold cumulative returns
portfolio_returns = pd.DataFrame(index=strategy_results[list(strategy_results.keys())[0]].index)

# Calculate cumulative returns for each stock
for ticker, df in strategy_results.items():
    portfolio_returns[ticker] = df['Cumulative_Return']

# Compute the average return across all stocks
portfolio_returns['Average'] = portfolio_returns.mean(axis=1)

# Calculate portfolio value over time
portfolio_returns['Portfolio_Value'] = initial_portfolio_value * portfolio_returns['Average']

# Plot the portfolio value over time
plt.figure(figsize=(12, 6))
plt.plot(portfolio_returns.index, portfolio_returns['Portfolio_Value'], label='Mean Reversion Strategy Portfolio')
plt.title('Portfolio Value Over Time')
plt.xlabel('Date')
plt.ylabel('Portfolio Value ($)')
plt.legend()
plt.show()

# Fetch SP500 data
def fetch_sp500_data(start, end):
    sp500 = yf.download('^GSPC', start=start, end=end)
    sp500['Daily_Return'] = sp500['Close'].pct_change()
    sp500['Cumulative_Return'] = (1 + sp500['Daily_Return']).cumprod()
    return sp500

sp500_data = fetch_sp500_data(start_date, end_date)

# Plot SP500 vs Portfolio
plt.figure(figsize=(12, 6))
plt.plot(portfolio_returns.index, portfolio_returns['Portfolio_Value'], label='Mean Reversion Strategy Portfolio')
plt.plot(sp500_data.index, sp500_data['Cumulative_Return'] * initial_portfolio_value, label='SP500', linestyle='--')
plt.title('Mean Reversion Portfolio vs SP500')
plt.xlabel('Date')
plt.ylabel('Portfolio Value ($)')
plt.legend()
plt.show()
```
